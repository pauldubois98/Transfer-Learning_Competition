```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
df = pd.read_csv('FFHQ/labelsFFHQ.csv')
df.columns
```

```python
n = 1000
nose_angle = np.array(df["nose_angle"][:n])
nose_area = np.array(df["nose_area"][:n])
nose_width = np.array(df["nose_width"][:n])
lips_area = np.array(df["lips_area"][:n])
eyes_ratio = np.array(df["eyes_ratio"][:n])

x =  np.arange(n)
a = np.argsort(nose_width)

plt.figure(facecolor='white')
plt.plot(x, nose_width[a]*180/3.1415)
plt.show()

imin = a[0]
imax = a[-1]

img = mpimg.imread("FFHQ/dataset/"+df.iloc[imin].filename)
imgplot = plt.imshow(img)
plt.axis('off')
plt.show()

img = mpimg.imread("FFHQ/dataset/"+df.iloc[imax].filename)
imgplot = plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
n = 1000
for col in df.columns[2:]:
    d = np.array(df[col][:n])
    x =  np.arange(n)
    a = np.argsort(d)


    fig, (ax0,ax1,ax2) = plt.subplots(1,3, figsize=(15,5), facecolor='white')
    fig.suptitle('(relative) '+col.replace('_', ' '), fontsize=16)
    ax0.plot(x, d[a]*180/3.1415)

    imin = a[0]
    imax = a[-1]

    img = mpimg.imread("FFHQ/dataset/"+df.iloc[imin].filename)
    ax1.imshow(img)
    ax1.axis('off')
    ax1.set_title('min')

    img = mpimg.imread("FFHQ/dataset/"+df.iloc[imax].filename)
    imgplot = ax2.imshow(img)
    ax2.axis('off')
    ax2.set_title('max')
    plt.savefig(col+'.png', )
    plt.show()
```
